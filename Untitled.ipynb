```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('ClientesBanco.csv', sep=',', encoding='ISO-8859-1')
df.drop('CLIENTNUM',axis='columns', inplace=True)
display(df)
```

```python
def grafico_rapido(df, coluna, numero_de_bar):
    plt.figure(figsize=(20,6))
    df.groupby(coluna).size().sort_values(ascending = False).head(numero_de_bar).plot(kind = 'bar', color = sns.color_palette("coolwarm"))
    plt.xlabel("Categoria")
    plt.ylabel("Quantidade")
    plt.title("xxx")
    plt.xticks(rotation=0)
    plt.show()
```

```python
def informacoes(df):
    print(' A media de meses como cliente é {:,.2f}'.format(df['Meses como Cliente'].mean()))
    print(' A media de produtos contratados {:,.2f}'.format(df['Produtos Contratados'].mean()))
    ###

    print(' O limite  medio é R${:,.2f}'.format(df['Limite'].mean()))
    print(' Com um consumo na media de {:,.2f}'.format(df['Limite Consumido'].mean()))
    #
    print(' A quantidade media de transações do cartão é {:,.2f}'.format(df['Qtde Transacoes 12m'].mean()))
```
